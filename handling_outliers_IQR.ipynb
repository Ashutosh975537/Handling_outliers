```python
df['placement_exam_marks'].describe()
x = df['placement_exam_marks']
```

```python
twenty_five_percentile_value = np.percentile(x,25)
seventy_five_percentile_value = np.percentile(x,75)
IQR = seventy_five_percentile_value - twenty_five_percentile_value
IQR
```

```python
upper_limit = seventy_five_percentile_value + ((1.5)*IQR)
lower_limit = twenty_five_percentile_value- ((1.5)*IQR)
```

```python
upper_limit
```

```python
lower_limit
```

# trimming

```python
mask = (df['placement_exam_marks'] <= upper_limit) & (df['placement_exam_marks'] >= lower_limit)
```

```python
df= df[mask]
```

# capping

```python
df = pd.read_csv('placement.csv')
```

```python
df['placement_exam_marks'] = np.where(df['placement_exam_marks'] > upper_limit , upper_limit , np.where(df['placement_exam_marks'] < lower_limit , lower_limit , df['placement_exam_marks']))
```
