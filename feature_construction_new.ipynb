```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LogisticRegression
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','Parch','SibSp','Survived']]
df.dropna(inplace=True)
```

```python
df.head()
```

```python
x = df.iloc[:,0:4]
```

```python
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x,y)
```

```python
np.mean(cross_val_score(lr,x,y,scoring = 'accuracy',cv=20))
```

# applying Feature construction

```python
x['Family_size'] = x['SibSp'] + x['Parch'] + 1 
```

```python
x.head()
```

```python
def categorize_family(size):
    if size == 1 :
        return 0 
    elif size >= 2 and size <= 4 :
        return 1 
    return 2 
```

```python
x['Family_size']
```

```python
x['Family_type'] = x['Family_size'].apply(categorize_family)
```

```python
x['Family_type']
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x
```

```python
np.mean(cross_val_score(lr,x,y,cv=20))
```
